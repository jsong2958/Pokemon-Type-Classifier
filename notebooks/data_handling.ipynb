```python
import os
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import DataFrame
import time
```

```python
# Dataset from https://www.kaggle.com/datasets/hlrhegemony/pokemon-image-dataset

destination = os.path.abspath("pokemon_images")

for root, dirs, files in os.walk("images"):
    for file in files:
        name = os.path.basename(root)
        file_path = os.path.join(root, file)
        new_file_name = name + "_" + file
        
        os.rename(file_path, os.path.join(destination, new_file_name))
        print(f"Renamed: {file} -> {new_file_name}")
        
```

```python
import re

def clean_string(s):
    s = s.lower()
    s = re.sub(r"[^\w\s-]", "", s)       
    s = re.sub(r"\s+", "-", s)            
    return s
```

```python
res = df.loc[df["type_1"].isnull()].values

for name in res:
    clean_name = name[0].split("_")[0]
    clean_name = clean_string(clean_name)
    url = "https://pokemondb.net/pokedex/" + clean_name
    types = []
    
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
    
        itypes = soup.find_all("a", class_="itype")
        for itype in itypes:
            if len(types) == 2:
                break
            if itype.text in types:
                break
            types.append(itype.text) 
            
        df.loc[df["img_name"] == name[0], ["type_1", "type_2"]] = [types[0], types[1] if len(types) == 2 else None]
        print(f"Changed {name[0]} to {types} in DataFrame")
        
    except Exception as e:
        print("Exception: ", e)
        
    time.sleep(2)
```
